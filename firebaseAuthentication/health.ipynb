```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
cema = pd.read_csv("C:/Users/USER/Downloads/cema_internship_task_2023.csv")

# Sort values by period
cema.sort_values(by='period', inplace=True)

# Plot Total Dewormed over period
plt.figure(figsize=(10, 6))
plt.plot(cema['period'], cema['Total Dewormed'], marker='o', linestyle='-')
plt.title('Total Dewormed Over Period')
plt.xlabel('Period')
plt.ylabel('Total Dewormed')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```
